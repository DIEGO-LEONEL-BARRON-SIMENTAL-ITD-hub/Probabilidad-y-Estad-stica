##**Ejercicio 3**

Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:


*   La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.

*   Cuantos estudiantes pesan más de 68 y menos de 72 kg.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math
```

```python
plt.figure(figsize=(6,4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500

# VARIABLE para Cálculo de la probabilidad de que un estudiante que pese menos de 72 kg
peso_limite = 72
# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 72 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_72kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese más de 72 kg es complementaria
prob_mas_72kg = 1 - prob_menos_72kg

# Cálculo del número de estudiantes que pesan menos de 72 kg
estudiantes_menos_72kg = total_estudiantes * prob_menos_72kg
print("Probabilidad de que un estudiante pese menos de 72 kg:", prob_menos_72kg)
print("Número de estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)

# Graficar la distribución normal
x = np.linspace (media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 72 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
peso_limite = 62 # Peso Limite
total_estudiantes = 500

# Calculamos la probabilidad de que un estudiante pese menos o igual a 62 kg
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese más de 62 kg es complementaria
prob_mas_62kg = 1 - prob_menos_62kg

# Calculamos el número aproximado de estudiantes que pesan más de 62 kg
estudiantes_mas_62kg = total_estudiantes * prob_mas_62kg
print("Probabilidad de que un estudiante pese más de 62 kg:", prob_mas_62kg)
print("Número aproximado de estudiantes que pesan más de 62 kg:", estudiantes_mas_62kg)

# Calcular la función de densidad de probabilidad de la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 62 Kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
# Calculamos el número de estudiantes que pesan menos de 62kg
estudiantes_menos_62kg = prob_menos_62kg * total_estudiantes
print("El número de estudiantes que pesan menos de 62kg: ", int(estudiantes_menos_62kg))

# Calculamos el número de estudiantes que pesan mas de 72kg
estudiantes_mas_72kg = prob_mas_72kg * total_estudiantes
print("El número de estudiantes que pesan más de 72kg: ", int(estudiantes_mas_72kg))

# Calculamos el número de personas en el rango de Personas que pesan entre 62 y 72kg
estudiantes_entre_62_72kg = total_estudiantes - (int(estudiantes_menos_62kg) + int(estudiantes_mas_72kg))
#Imprimimos el número de estudiantes en el rango
print("El número de estudiantes que pesan entre 62kg y 72kg: ",estudiantes_entre_62_72kg)

peso_lim_inferior = 62
peso_lim_superior = 72

# Calcular la función de densidad de probabilidad de la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=((x >= peso_lim_inferior)&(x <= peso_lim_superior)), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante tenga un peso entre 62 Kg y 72kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
