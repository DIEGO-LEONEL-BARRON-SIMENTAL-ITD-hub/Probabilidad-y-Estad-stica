```python
import math as mt #para utilizar factorial
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes= ["Alan", "Sofía", "Mateo", "Valeria", "Bruno", "Lucía", "Diego", "Camila", "Andrés", "Natalia"]
n= len(asistentes)
r = 4 # ¿Cuántas preguntas?

Pr = mt.factorial(n) / (mt.factorial(n-r))

print("Existen ", Pr , " posibles permutaciones sin repetición del total de ",
n," asistentes ", " para calificar ", r , "preguntas")

print("Aquí las permutaciones")
permutaciones_asistentes = list(permutations(asistentes, r))

# Imprime las permutaciones
for permutacion in permutaciones_asistentes:
  print(permutacion)
```
