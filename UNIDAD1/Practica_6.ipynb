```python
from statistics import *
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
import math
from statistics import stdev
```

```python
# Cargar el archivo CSV en una variable
autos=pd.read_csv("autos2.csv",encoding="ISO-8859-1")
```

```python
#mostrar los primeros 5 registros
autos.head()
```

```python
#Mostrar el precio máximo de los autos
autos.price.max()
```

```python
#Mostrar el precio mínimo de los autos
autos.price.min()
```

```python
#Calcular media, mediana y moda y guardarlos en variables
media=autos.price.mean()
mediana=autos.price.median()
moda=autos.price.mode()

print(media)
print(mediana)
print(moda)
```

```python
#Calcular la desviación estándar de los precios de los autos
desvEst=autos.price.std()
print(desvEst)
```

```python
# Declarar una variable para mejorar la calidad de vida
precio=autos.price
```

```python
print(precio.describe(include='all'))
```

```python
#Calcular el número de intervalos utilizando la ley de sturges
n=len(precio)
K=int(1+math.log2(n))
print("Valor de n = ",n,"\nNumero de intervalos = ", K)
```

```python
#Declarar una moda útil
modamax=precio.value_counts().idxmax()
print(modamax)
```

```python
plt.figure(figsize=(8,6))
plt.hist(precio, bins=K, alpha=0.2, color='b', edgecolor="black")
plt.axvline(media,color='b', linestyle='dashed',linewidth=2,label=f'Media={media:.2f}')
plt.axvline(mediana,color='r', linestyle='dashed',linewidth=2,label=f'Mediana={mediana:.2f}')
plt.axvline(modamax,color='y', linestyle='dashed',linewidth=2,label=f'Moda={modamax:.2f}')
plt.axvline(media-desvEst,color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={desvEst:.2f}')
plt.axvline(media+desvEst,color='black', linestyle='dashed', linewidth=2)

plt.xlabel('Precios')
plt.ylabel('Frecuencia')
plt.title("Precios de los Autos")
plt.legend()
plt.show()
```

En la gráfica se puede observar la tendencia a la procción de autos de uso cotidiano y económico, dado que las medidas de tendencia central marcan valores inferiores a los $20,000, siendo así que los autos de lujo son poco comunes.

```python
sns.boxplot(x='body-style',y='price',data=autos)
plt.title("Relación tipo de auto-precio")
plt.ylabel("Precio")
plt.xlabel("Tipo de auto")
```

Los autos deportivos o de lujo como al parecer hardtop y convertibles tienden a precios elevados debido a su diseño especializado, mientras que hatchback y sedan son más asequibles y comunes. Así como hay valores extraordinarios en sedan y wagon que indican que, si bien su mercado fuerte son los autos más económicos, también tienen modelos de lujo

```python
sns.boxplot(x='drive-wheels',y='price',data=autos)
plt.title("Relación Tipo de tracción-precio")
plt.ylabel("Precio")
plt.xlabel("Tipo de tracción")
```

Se puede observar un claro estandar en la tracción trasera de los autos, así como que los autos con tracción delantera o 4x4 no están dentro de los más caros del mercado

```python
sns.boxplot(x='engine-location',y='price',data=autos)
plt.title("Relación Ubicación del motor-precio")
plt.ylabel("Precio")
plt.xlabel("Ubicación del motor")
```

Aquí se observa claramente la preferencia y el estándar del motor delantero, así como que los autos que tienen el motor en la parte de atrás, tienden a un precio más elevado
