##**9. Diagrama de Cajas y bigotes**
1. Las líneas se usan para indicar la variabilidad fuera de los cuartiles superior e inferior.
2. Los valores atípicos se representan a veces como puntos individuales que están en la línea con los bigotes.
3. **Si los datos son simétricos, la mediana estará en el centro de la caja.**
4. Si los datos están **sesgados**, la medianaestará más cerca de la parte superior o inferior de la caja.
5. Los extremos de arriba y abajo de la caja indican los **cuantiles, o percentiles, 25 y 75**. Estos dos cuantiles también se conocen como cuartiles, porque separan cuantos (25%) de los datos. La longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercuartílico (IQR).
6. Los diagramas de caja-bigotes son muy útiles para comparar dos o más conjuntos de datos.


```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio=round(random.uniform(40,93),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
plt.boxplot(fabrica1)
plt.ylabel('Producción en mts.')
plt.title('Producción de la fabrica 1')
plt.show()
```

```python
plt.boxplot(fabrica2)
plt.ylabel('Producción en mts.')
plt.title('Producción de la fabrica 1')
plt.show()
```

```python
fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(9,4))
bplot1= ax1.boxplot(fabrica1,
                    vert=True, #Alineación vertical
                    patch_artist=True) #Rellenar con color
ax1.set_title('Producción fabrica 1')

bplot2= ax2.boxplot(fabrica2,
                    vert=True, #Alineación vertical
                    patch_artist=True) #Rellenar con color
ax2.set_title('Producción fabrica 2')

plt.show()
```

```python
#Insertamos un valor de 150 a la fabrica 1,
#Sería un valor atípico

fabrica1.append(150)
print(fabrica1)
```

```python
fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(9,4))
bplot1= ax1.boxplot(fabrica1,
                    vert=True, #Alineación vertical
                    patch_artist=True) #Rellenar con color
ax1.set_title('Producción fabrica 1')

bplot2= ax2.boxplot(fabrica2,
                    vert=True, #Alineación vertical
                    patch_artist=True) #Rellenar con color
ax2.set_title('Producción fabrica 2')

plt.show()
```

```python
# Agregamos ax1.set_ylim(0, 160) para garantizar que ambas usen la misma escala vertical, facilitando la comparación visual

fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(9,4))
bplot1= ax1.boxplot(fabrica1,
                    vert=True, #Alineación vertical
                    patch_artist=True) #Rellenar con color
ax1.set_title('Producción fabrica 1')
ax1.set_ylim(30,160)
bplot2= ax2.boxplot(fabrica2,
                    vert=True, #Alineación vertical
                    patch_artist=True) #Rellenar con color
ax2.set_title('Producción fabrica 2')
ax2.set_ylim(30,160)
plt.show()
```
