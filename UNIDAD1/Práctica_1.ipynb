```python
from statistics import *
```

```python
# Lista de calificaciónes del grupo a de probabilidad y estadística
grupoa=[70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
```

```python
# La función type nos devuelve el tipo de dato de la variable
type(grupoa)
```

```python
# Sacar la media de las calificaciones con una fórmula
promedio = sum(grupoa)/len(grupoa)
print(promedio)
```

```python
#Utilizar la función mean que está en la librería en la librería statistics para sacar el promedio
promedio2 = mean(grupoa)
print(promedio2)
```

```python
# Determinar el valor de la mediana con la función median
mediana = median(grupoa)
print(mediana)
```

```python
# Determinar el valor de la moda con la función mode
moda = mode(grupoa)
print(moda)
```

```python
# En caso de que exista más de una moda, se puede emplear la función multimode para sacar múltiples modas que puedan existir
moda2 = multimode(grupoa)
print(moda2)
```

## Tiempos de 12 alumnos en 100 metros

```python
#Definimos una variable llamada tiempos100mts de tipo lista, que contiene los tiempos que tardaron los 10 alumnos en correr 100 metros
tiempos100mts=[15.10,17.2,14.69,13.27,22.15,18.71,17.15,20.3,15.1,17.2]
```

# Ejemplo 3: Histogramas de 2 grupos


```python
# Librería para generar números aleatorios
import random
```

```python
fabrica1= []
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2= []
for i in range(100):
  numero_aleatorio=round(random.uniform(70,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
# Librería de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Producción diaria de la Fábrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Producción diaria de la Fábrica 2")
```

```python
print("Valores de la fabrica 1")
print("La media es =",mean(fabrica1))
print("La mediana es =",median(fabrica1))
print("La moda es =",mode(fabrica1))
```

```python
print("Valores de la fabrica 2")
print("La media es =",mean(fabrica2))
print("La mediana es =",median(fabrica2))
print("La moda es =",mode(fabrica2))
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
