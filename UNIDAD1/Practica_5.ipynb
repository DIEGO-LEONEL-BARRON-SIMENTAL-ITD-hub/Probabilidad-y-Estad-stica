```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
# Cargar el archivo CSV en una variable llamada covid
covid=pd.read_csv("20-06-30-COVID19MEXICO - codificado.csv",encoding="ISO-8859-1")
```

```python
# Mostramos los primeros 5 registros del archivo
covid.head()
```

```python
#Mostrar la información del archivo.
#Muestra el número de registros, tipos de datos de la columna, valores nulos, memoria utilizada, etc.
covid.info()
```

```python
# mostrar la edad máxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
#mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
#Dibujar el histograma del atributo "EDAD" con bins = 3
#Calcular la media y guardarla en una variable para mostrarla en el histograma
media = covid.EDAD.mean()
plt.pyplot.hist(covid.EDAD, bins =11)
plt.pyplot.title("Histograma Edad Pruebas COVID")
plt.pyplot.axvline(media,color='y', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.pyplot.xlabel("Edad")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Dibuja el histograma de la columna sexo, esto permite comparar cuantos hay de cada uno
plt.pyplot.hist(covid.SEXO, bins=3)

#establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma Sexo Pruebas COVID")
plt.pyplot.xlabel("Sexo")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Dibuja el histograma del atributo "Tipo Paciente", esto permite comparar cuantos hay de cada grupo
plt.pyplot.hist(covid.TIPO_PACIENTE, bins=3)

#Establece las etiquetas x/y y muestra el titulo
plt.pyplot.title("Histograma de tipos de Pacientes en Pruebas COVID")
plt.pyplot.xlabel("Tipo de paciente")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Crea un diagrama de cajas para visualizar la distribución de los rangos de edad por sexo
sns.boxplot(x='SEXO', y='EDAD', data=covid)
```

```python
#Crea un diagrama de cajas para visualizar la distribución de los rangos de edad por tipo de paciente
sns.boxplot(x='TIPO_PACIENTE', y='EDAD', data=covid)
```

```python
#Crea un diagrama de cajas para visualizar la distribución de los rangos de edad por resultado
sns.boxplot(x='RESULTADO', y='EDAD', data=covid)
```
