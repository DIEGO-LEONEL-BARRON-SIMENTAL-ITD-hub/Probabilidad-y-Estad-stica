```python
import math
import numpy as np
import matplotlib.pyplot as plt

x= [5,6,7,8,9,10] #Valores posibles de la variable aleatoria
p= [0.05,0.10,0.20,0.30,0.25,0.10] #Probabilidades correspondientes

# Calcular la función acumulada
F= np.cumsum(p)

# Mostrar resultados
for xi, fi in zip(x,F):
  print(f"F({xi}) = {fi:.2f}")

#Graficar
plt.step(x,F, where='post')
plt.title("Función de Distribución Acumulada (FDA)")
plt.xlabel("Calificaión")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
