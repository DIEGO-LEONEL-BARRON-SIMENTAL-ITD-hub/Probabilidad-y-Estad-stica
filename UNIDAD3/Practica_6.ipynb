```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Variable aleatoria y sus probabilidades
x= np.array([0,1,2,3,4,5])
p= np.array([0.10,0.25,0.30,0.20,0.10,0.05])

print(f"La suma de las probabilidades es: {p.sum()}")

# Valor esperado
esperanza = np.sum(x*p)

#Esperanza al cuadrado
esperanza_cuadrado = np.sum(x**2 *p)

# Varianza y desviación estándar
varianza = esperanza_cuadrado - esperanza **2
desviacion = np.sqrt(varianza)

# Mostrar los resultados en consola
print(f"E(X) = {esperanza:.2f}")
print(f"E(X²) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación Estándar = {desviacion:.2f}")

# Crear la gráfica 'stem' (tallos y hojas)
plt.stem(x, p, basefmt=" ", linefmt='b-',markerfmt='bo')
plt.title("Distribución de la probabilidad Discreta\ncon Media y Desviación Estándar")
plt.xlabel("Número de celulas anormales por muestra")
plt.ylabel("P(X)")
plt.grid(True, linestyle='--', alpha=0.6)

# Linea de la media
plt.axvline(x=esperanza, color='red', linestyle='--', label=f"Media (E(X)) = {esperanza:.2f}")

# Rango de ±1 desviación estándar
plt.axvspan(esperanza-desviacion, esperanza+desviacion, color='red',alpha=0.2,label=f"±1 Desv. Estándar (σ = {desviacion:.2f})")

# Leyenda y límites
plt.legend()
plt.ylim(0, max(p) + 0.05)
plt.xticks(x)

# Gráfica
plt.show()
```

```python
#Función acumulada
F = np.cumsum(p)

# Resultados
for xi, fi in zip(x,F):
  print(f"F({xi}) = {fi:.2f}")

#Grafica
plt.step(x,F, where='post')
plt.title("Función de Distribución Acumulada Celulas anormales por muestra")
plt.xlabel("Celulas anormales por muestra")
plt.ylabel("Probabilidad acumulada")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
