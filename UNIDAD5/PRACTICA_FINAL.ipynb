```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
import numpy as np
from scipy.stats import pearsonr
```

```python
# Cargar los documentos en variables
df1= pd.read_csv("grupo1.csv")
df2= pd.read_csv("grupo2.csv")
df3=  pd.read_csv("grupo3.csv")
```

```python
# Solicitamos los primeros 5 registros
df1.head()
```

```python
df1.describe()
```

```python
# Cargar datos en Variables
cal1=df1["Calificacion"]
hrs1=df1["HorasEstudio"]
```

```python
# Graficamos Histograma

plt.hist(cal1, bins=10, alpha=0.8, color='r')
plt.title("Calificaciones Grupo 1")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Dibujar el Diagrama de Dispersión
sns.scatterplot(x = hrs1, y = cal1, color='red')
plt.suptitle('Grupo 1')
plt.title('Diagrama de Dispersión de Calificación vs. Horas de Estudio')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```

```python
# Dibujar gráfico de Dispersión con Linea de Regresión
sns.set(style="whitegrid")
sns.regplot(x= hrs1, y=cal1, scatter=True, color='red')
plt.title('Diagrama de Dispersión Con Línea de Regresión')
plt.suptitle('Grupo 1')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaión')
plt.grid(True)
plt.show()
```

```python
# Crear Arrays
arCal1 = np.array(cal1)
arHrs1 = np.array(hrs1)

# Crear y entrenar el modelo de regresión lineal
modelo1 = LinearRegression()
modelo1.fit(arHrs1.reshape(-1, 1), arCal1)

# Predecir la calificación obtenida para un tiempo de estudio de 15 horas.
tiempo_prediccion1 = np.array([[15]])
calificacion_predicha1 = modelo1.predict(tiempo_prediccion1)

print("La calificación esperada para un tiempo de 15 horas de estudio es:", calificacion_predicha1[0])
print("Valor de la intercepción:", modelo1.intercept_)
print("Valor de la pendiente:", modelo1.coef_)

# Crear el diagrama de dispersión con la línea de regresión
plt.figure(figsize=(6, 5))
plt.scatter(arHrs1, arCal1, color='blue', label='Datos')
plt.plot(arHrs1, modelo1.predict(arHrs1.reshape(-1, 1)), color='red', label='Línea de regresión')
plt.scatter(tiempo_prediccion1, calificacion_predicha1, color='green', marker='x', s=100,
            label='Predicción (15h, {0:.2f})'.format(calificacion_predicha1[0]))
plt.title('Pronóstico de la calificación a partir del tiempo de estudio')
plt.xlabel('Tiempo de estudio (Horas)')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcular el coeficiente de correlación de Pearsonr
coeficiente_correlacion, p_valor = pearsonr(hrs1, cal1)

print("Coeficiente de correlación de Pearsonr: ", coeficiente_correlacion)
print("Valor p: ", p_valor)
```

#**PREGUNTAS GRUPO 1**
###1.  Explica: ¿Cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?
La pendiente en este caso es muy pequeña, y se puede observar que incluso en negativa, por lo que parece que a éste grupo estudiar más disminuye su calificación aunque con un impacto realmente minúsculo.
###2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?
Dado que en este grupo los puntos se encuentran muy dispersos, no es de extrañarse que su coeficiente de correlación sea de -0.06 aproximadamente, por lo tanto, se podría decir que ambos factores en este grupo presentan una relación muy debil o casi nula.
###3. ¿El modelo de regresión lineal es estadísticamente significativo?
Al parecer, el modelo de regresión lineal no resulta significativo, dado que la pendiente de la recta es muy poco pronunciada, indicando que no hay una relación lineal fuerte.
###4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?
El modelo no es muy confiable para predecir calificaciones a partir del tiempo de estudio, porque los puntos están muy dispersos alrededor de la recta y hay mucha variabilidad no explicada.
