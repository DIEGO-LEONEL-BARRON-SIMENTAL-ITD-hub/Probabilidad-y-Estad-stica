```python
from scipy.stats import pearsonr

# Datos de entrada
publicidad = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Calcular el coeficiente de correlación de Pearsonr
coeficiente_correlacion, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlación de Pearsonr: ", coeficiente_correlacion)
print("Valor p: ", p_valor)
```
